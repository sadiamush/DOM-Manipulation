## Logical Operations

Logical operations allow combining multiple boolean expressions and are crucial for decision-making processes in programming. Here are the basics:


1. **`and`**
   - **Purpose:** Checks if all conditions are `True`.
   - **True Example:** `(3 < 5) and (-1 == -1)` evaluates to `True` because both conditions are `True`.
   - **False Example:** `(7 == 8) and (12 > 2)` evaluates to `False` because the first condition is `False`.

2. **`or`**
   - **Purpose:** Checks if at least one condition is `True`.
   - **True Example:** `(0 == 1) or (10 <= 15)` evaluates to `True` because the second condition is `True`.
   - **False Example:** `(0 > 5) or (6 == 3)` evaluates to `False` because both conditions are `False`.

3. **`not`**
   - **Purpose:** Returns `True` if the condition is `False`, and vice versa.
   - **True Example:** `not False` evaluates to `True`.
   - **False Example:** `not (1 > 0)` evaluates to `False` because `(1 > 0)` is `True`.

## 1️⃣ For loop:


Python's for loops are a powerful tool for automating repetitive tasks, making your code more efficient and easier to manage.

- **For loops in Python** iterate over a sequence (such as a list, tuple, string, or range).

- For loops are beneficial for performing repetitive tasks efficiently.
- They can handle **both numerical and non-numerical data** types.

*A for-loop has **three** parts:*

**Loop Variable**- the variable that changes after each iteration of the loop

**Range** - Which values the loop variable takes on in each iteration

**Code Body** - the block of code that gets repeated at each iteration









### Range in for loop

- **`range()`** sets the number of times your code will loop.
- It **provides the values** that the for-loop will iterate over.
- **Basic usage**: `for i in range(n)` loops from 0 to `n-1`.
- You can **customize the range**:
  - Start from a number other than 0: `range(start, stop)`
  - Specify a step: `range(start, stop, step)`
- Useful for looping over **specific sequences** like:
  - A range of numbers (e.g., from 5 to 10)
  - Even or odd numbers (e.g., `range(0, 10, 2)` for evens).

### **Infinite Loops**


-    Don’t actually write code that will never stop.

-    An infinite loop is when your code gets stuck in an endless cycle of repeating code and never finishes.

-    It’s pretty hard to get a Python for-loop stuck in an infinite loop, but we need to be especially careful with while-loops.

-    The key to prevent an infinite loop is to look for conditions that will never be False

````python
# The boolean value True will never be False
while True:
    print("I am always True")
````

````python
# i keeps decreasing and will always be less than 1
i = 0
while i < 1:
    i = i - 1
````

````python
# Since two boolean values are always either equal or not equal this
# condition will never be false
a = True
b = False
while (a == b) or (a != b):
    print("Can't make up your mind?")
````

### Break Statement

````python
# Example of break statement
for i in range(1, 10):
    if i == 5:
        print("Breaking the loop as i equals 5")
        break  # Exit the loop when i is 5
    print(f"i is: {i}")
````

### **Continue Statement:**
The continue statement skips the current iteration and moves to the next iteration. When the continue statement is executed, the loop doesn't terminate but jumps to the next iteration, bypassing any code that follows it within the Python loop.


Continue statement cannot be used effectively without an if statement or a conditional check. This is because the purpose of continue is to conditionally skip the rest of the current loop iteration and move to the next one. Without an if statement or a condition, continue would simply always skip the rest of the loop, which would make it redundant.



What will happen if you place a `continue` statement in the middle of a `while` loop that has an incrementing counter inside the loop?

````python
counter= 0
while counter<5:
  if counter==2:
    continue
  print("Gksdkjd",counter)
  counter=counter+1
print("last update", counter)
````

- **Initialization**: The list `numbers` is initialized with values from 1 to 10.
- **Iteration**: The `for` loop iterates over each number in the list.
- **Condition**: The loop checks if the current number is even (`number % 2 == 0`).
- **Continue**: If the number is even, `continue` is executed, skipping the current iteration.
- **Processing**: If the number is odd, it is printed to the console.




### `pass`
- **Does nothing**: It's a placeholder statement.
- **Used when a statement is syntactically required but you don't want to execute any code**.
- **Does not affect the flow**: The loop or block continues to the next statement.


- **`pass`**: Does nothing, and the loop continues without skipping any iterations.
- **`continue`**: Skips the current iteration and moves to the next one.

````python
for number in range(1, 6):
    if number == 3:
        pass  # Does nothing, just a placeholder
    print(f"Number is: {number}")
````

A **placeholder statement** in programming refers to a statement that is syntactically required but doesn't actually perform any action. It's used to fill in code blocks where a statement is needed to avoid errors, but no logic needs to be executed at that point.

In Python, the `pass` statement is the most common placeholder. It allows you to write code that is **syntactically correct** while deferring the actual logic until later.

### Why Use a Placeholder Statement (`pass`)?

1. **Avoid Syntax Errors**: Python requires indented blocks of code (like after an `if`, `for`, or function definition), and using `pass` ensures the code block is valid even if it's empty.
2. **Writing Skeleton Code**: When designing a program, you might outline the structure first and fill in the details later. `pass` lets you create placeholders where logic will eventually be implemented.

### Example:

```python
def my_function():
    pass  # Placeholder for future implementation

if True:
    pass  # Placeholder, no action here yet

for i in range(5):
    pass  # Loop exists, but doesn't do anything yet
```

In each of these examples, `pass` allows the code to run without producing an error, but it doesn't do anything. It's like saying, "I'll come back and implement this part later."
